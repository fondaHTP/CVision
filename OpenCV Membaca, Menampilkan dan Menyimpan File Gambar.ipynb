```python
import cv2
```

```python
# Aktifkan path
gbrpath = cv2.imread("D:\\Latihan C.Vision")
# Baca File gambar
gbr = cv2.imread('Anak Kucing.jpg')
```

```python
# Tampilkan bambar
frame='gbr'
cv2.imshow('frame',gbr)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Tampilkan gambar dengan grafik matplotlib
import matplotlib.pyplot as plt
plt.imshow(gbr)
```

```python
# Rubah gambar matplotlib dari bgr opencv ke rgb matplotlib
gbrbaru=cv2.cvtColor(gbr,cv2.COLOR_BGR2RGB)
plt.imshow(gbrbaru)
```

```python
gbr = cv2.imread('Anak Kucing.jpg',cv2.IMREAD_GRAYSCALE)
frame='gbrlagi'
cv2.imshow('frame',gbr)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
gbr.shape
```

```python
# Tampilkan gambar grey dengan grafik matplotlib
plt.imshow(gbr)
```

```python
# Rubah gambar matplotlib dari grey opencv ke rgb matplotlib
gbrgrey=cv2.cvtColor(gbr,cv2.COLOR_GRAY2BGR)
plt.imshow(gbrgrey)
```
