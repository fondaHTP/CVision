```python
import cv2
```

```python
gbr = cv2.imread('Cat.jpg')
```

```python
# Cek gambar apakah sudah benar?
type(gbr)
```

```python
# Lihat ukuran Gambar
gbr.shape
```

```python
# Tampilkan bambar
Frame='gbr'
cv2.imshow('Frame',gbr)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Tampilkan gambar dengan grafik matplotlib
import matplotlib.pyplot as plt
plt.imshow(gbr)
```

```python
# Rubah gambar matplotlib dari bgr opencv ke rgb matplotlib
gbrbaru=cv2.cvtColor(gbr,cv2.COLOR_BGR2RGB)
plt.imshow(gbrbaru)
```
