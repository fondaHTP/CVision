# EQUALISASI HISTOGRAM 
### Merupakan suatu cara yang bertujuan untuk memperoleh histogram yang intensitasnya terdistribusi secara seragam pada citra. Pendekatan yang dilakukan adalah untuk mendapatkan aras keabuan yang lebih luas pada daerah yang memiliki banyak piksel  dan mempersempit aras keabuan pada daerah yang berpiksel sedikit. Efeknya dapat digunakan untuk meningkatkan kontras secara menyeluruh.

## openCV fungsi:  cv2.equalizeHist(src[, dst])

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

gambar = cv2.imread('anak kucing grey.jpg',0)

# lihat ukuran gambar (jika tidak ada ukuran araynya berarti gambar tidak ada di folder)
gambar.shape
```

```python
# Tampil gambar
cv2.imshow('gambar',gambar)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# lakukan perhitungan untuk histogram
hist_asli=cv2.calcHist([gambar],[0],None,[256],[0,256])
plt.plot(hist_asli)
```

```python
gambar_equal=cv2.equalizeHist(gambar)
rgb_ekualisasi=cv2.cvtColor(gambar_equal,cv2.COLOR_GRAY2RGB)
plt.imshow(rgb_ekualisasi)
plt.title('Gambar Equalisasi')
plt.show()

hist_equalisasi=cv2.calcHist([gambar_equal],[0],None,[256],[0,256])
plt.title('Grafik Histogram Equalisasi')
plt.plot(hist_equalisasi)
```

```python
# concatenate untuk menampil gambar berdampingan
gambar_conconate_horiz=np.concatenate((gambar,gambar_equal),axis=1)
# axis=1 untuk tampil gambar horizontal dan axis=0 untuk tampil gambar secara vertikal


# tampilkan gambar secara horizontal
cv2.imshow('Perbandingan Gambar Equalisasi',gambar_conconate_horiz)
cv2.waitKey()
cv2.destroyAllWindows()
```

# Ekualisasi Gambar Warna

```python
gambar=cv2.imread('anak kucing.jpg')

# rubah gambar gbr openCV ke rgb matplotlib
gambar_bgr=cv2.cvtColor(gambar,cv2.COLOR_BGR2RGB)

# Tampilkan gambar
gambar_plot = plt.imshow(gambar_bgr)
plt.title('Gambar Asli')
plt.show()

# Pisahkan warna pada gambar
pisah_gambar=cv2.split(gambar)
warna=('b','g','r')

# lakukan perlukangan untuk setiap warna r,g dan b untuk histogram
for i, color in zip(pisah_gambar,warna):
    histr1=cv2.calcHist([i],[0],None,[256],[0,256])
    plt.plot(histr1,color=color)
    plt.xlim([0,256])
plt.title('Histogram Gambar Asli')
plt.show()

# pisah bgr image ke 3 chanel
b,g,r = cv2.split(gambar)

b=cv2.equalizeHist(b);
g=cv2.equalizeHist(g);
r=cv2.equalizeHist(r);

# Gabung chanel 
bgr_gabung=cv2.merge([b,g,r])

# Rubah gambar hasil equalisasi gbr openCV ke rgb matplotlib
rgb_equalisasi=cv2.cvtColor(bgr_gabung,cv2.COLOR_BGR2RGB)

# Tampilkan gambar
plt.imshow(rgb_equalisasi)
plt.title('Gambar Equalisasi')
plt.show()

# Histogram Equalisasi
pisah_gambar=cv2.split(rgb_equalisasi)
# lakukan perlukangan untuk setiap warna r,g dan b untuk histogram
for i, color in zip(pisah_gambar,warna):
    histr2=cv2.calcHist([i],[0],None,[256],[0,256])
    plt.plot(histr2, color=color)
    plt.xlim([0,256])

plt.title('Histogram Gambar Equalisasi')
plt.show()
```

```python
pisah_gambar=cv2.split(gambar)
warna=('b','g','r')
gambar_channel=[]
for i, color in zip(pisah_gambar,warna):
    gambar_channel.append(cv2.equalizeHist(i))
    gambar_equalisasi=cv2.merge(gambar_channel)

# tampilkan gambar secara horizontal 
gambar_horiz=np.concatenate((gambar, gambar_equalisasi),axis=1)
cv2.imshow('Perbandingan Gambar Asli dan Gambar Equalisasi',gambar_horiz)
cv2.waitKey()
cv2.destroyAllWindows()
```
