# DETEKSI TEPI (EDGE DETECTION) CANNY 

## Detektor tepi Canny menggunakan filter Gaussian dua dimensi untuk menghilangkan noise, kemudian menerapkan deteksi tepi Sobel dengan penekanan non-maksimum untuk memilih nilai rasio maksimum antara gradien x dan y pada titik piksel mana pun dan, akhirnya, menerapkan ambang batas tepi untuk mendeteksi ada atau tidaknya tepi.


### Edge = Canny(image, edges, threshold1, threshold2)
- image Objek yang mewakili sumber (gambar input) untuk operasi ini.
- edge Objek mewakili tujuan (tepi) untuk operasi ini.
- threshold1 Variabel tipe double yang mewakili ambang pertama untuk prosedur histeresis.
- threshold2 Variabel tipe double yang mewakili ambang kedua untuk prosedur histeresis.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('Kuda.jpg')
```

```python
plt.imshow(img)
```

```python
edges = cv2.Canny(image=img, threshold1=127, threshold2=127)
```

```python
plt.imshow(edges)
```

```python
edges = cv2.Canny(image=img, threshold1=0, threshold2=255)
```

```python
plt.imshow(edges)
```

## Choosing Thresholds

https://stackoverflow.com/questions/25125670/best-value-for-threshold-in-canny

```python
# hitung nilai pixel median
med_val = np.median(img) 
```

```python
# Batas bawah adalah 0 atau 70% dari nilai median, mana saja yang lebih tinggi
lower = int(max(0, 0.7* med_val))
```

```python
# Batas atas adalah 255 atau 30% di atas nilai median, mana saja yang lebih rendah
upper = int(min(255,1.3 * med_val))
```

```python
edges = cv2.Canny(image=img, threshold1=lower , threshold2=upper)
```

```python
plt.imshow(edges)
```

Terkadang membantu mengaburkan gambar terlebih dahulu, tepi kecil tidak di ambil.

```python
blurred_img = cv2.blur(img,ksize=(5,5))
```

```python
edges = cv2.Canny(image=blurred_img, threshold1=lower , threshold2=upper)
```

```python
plt.imshow(edges)
```
